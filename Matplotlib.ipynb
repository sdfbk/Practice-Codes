```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x= [1,3,5,7,9,11]
y = [11,14,15,19,24,25]
plt.figure(figsize= [10,7])
plt.plot(x,y,marker = "*")
plt.grid(True)
plt.title("Revenue and Time",color="yellow")
plt.xlabel("Time",color = "red")
plt.ylabel("Revenue",color = "green")
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
plt.show()
```

```python
a=np.arange(1,20)
y1=np.sqrt(a)
y2=np.log(a)
```

```python
plt.figure(figsize=[5,5])
plt.plot(a,y1,color = "red",label = "Square Root")
plt.plot(a,y2,color = "green",label = "Log")
plt.legend()
plt.grid(True)
plt.title("Sqrt and log",color = "blue")
plt.show()
```
