```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso
from sklearn.metrics import r2_score,accuracy_score
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

```python
df = df.drop(['normalized-losses','horsepower'],axis = 'columns')
```

```python
df
```

```python
df_cat = df.select_dtypes('object')
df_num = df.select_dtypes(include = ['int','float'])
for col in df_cat:
    le = LabelEncoder()
    df_cat[col] = le.fit_transform(df_cat[col])
df = pd.concat([df_cat,df_num],axis = 1)
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.price
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.20,random_state = 1)
```

```python
for i in range(1,31):
    l2 = Ridge(alpha = i)
    l2.fit(xtrain,ytrain)
    yp = l2.predict(xtest)
    print("Training accuracy : ",r2_score(l2.predict(xtrain),ytrain),"\t","Testing accuracy : ",r2_score(ytest,yp))
```

```python
l2 = Ridge(alpha = 8)
l2.fit(xtrain,ytrain)
yp = l2.predict(xtest)
print("Training accuracy : ",r2_score(l2.predict(xtrain),ytrain),"\t","Testing accuracy : ",r2_score(ytest,yp))
```

```python
for i in range(1,31):
    l1 = Lasso(alpha = i)
    l1.fit(xtrain,ytrain)
    yp = l1.predict(xtest)
    print("Training accuracy : ",r2_score(l1.predict(xtrain),ytrain),"\t","Testing accuracy : ",r2_score(ytest,yp))
```

```python
l1 = Lasso(alpha = 10)
l1.fit(xtrain,ytrain)
yp = l1.predict(xtest)
print("Training accuracy : ",r2_score(l1.predict(xtrain),ytrain),"\t","Testing accuracy : ",r2_score(ytest,yp))
```
