```python
html_doc="""
<html>
<title>This is title</title>
<body>
<h1>This is heading tag</h1>
<ul class="Clothes">
<li>Shirts</li>
<li>T-Shirts</li>
<li>Shorts</li>
</ul>
<ul class="brands">
<li>Zara</li>
<li>H&M</li>
<li>Puma</li>
</ul>
</body>
</html>
"""
```

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
html_=bs(html_doc,"html.parser")
print(html_)
print(html_.prettify())
print(html_.find("h1"))
header_=html_.find("h1")
print(header_.text)
```

```python
list1=[]
soup=bs(html_doc,"html.parser")
ul_=soup.find("ul",class_="Clothes")
li_=ul_.find_all("li")
for i in li_:
    list1.append(i.text)
print(list1)
```

```python
list2=[]
soup=bs(html_doc,"html.parser")
ul_=soup.find("ul",class_="brands")
li_=ul_.find_all("li")
for i1 in li_:
    list2.append(i1.text)
print(list2)
```

```python
dict1={"Clothes":list1,"Brands":list2}
print(dict1)
```

```python
import bs4
from bs4 import BeautifulSoup as bs
import requests
a=requests.get("https://www.javatpoint.com/","html.parser")
soup=bs(a.text)
print(soup.prettify())
```

```python
list1=[]
div_tags=soup.find("div",class_="ddsmoothmenu")
li_tags=div_tags.find_all("li")
for lists in li_tags:
    list1.append(lists.text)
print(list1)
```

```python
list2=[]
div_tags=soup.find("div",class_="ddsmoothmenu")
li_tags=div_tags.find_all("li")
for lists in li_tags:
    a_tags=lists.find("a")
    list2.append(a_tags.get("href"))
print(list2)
```

```python
list3=[]
div_tags=soup.find("div",class_="ddsmoothmenu")
li_tags=div_tags.find_all("li")
for li in li_tags:
    a_tags=li.find("a")
    img_tags=a_tags.find("img")
    list3.append(img_tags.get("src"))
print(list3)
```

```python
javatpoint={"Course":list1,"Course link":list2,"Course image links":list3}
print(javatpoint)
```

```python
import pandas as pd
df=pd.DataFrame(javatpoint)
df
```

```python
file_csv=df.to_csv("javatpoint.csv",index=True)
```
