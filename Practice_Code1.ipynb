```python
def calci(x1,x2,opr):
    if opr=="+":
        return f"Addition: {x1+x2}"
    elif opr=="-":
        return f"Subtraction: {x1-x2}"
    elif opr=="*":
        return f"Multiplication: {x1*x2}"
    elif opr=="/":
        return f"Division: {x1/x2}"
    elif opr=="%":
        return f"Modulus: {x1%x2}"
    elif opr=="**":
        return f"Exponents: {x1**x2}"
    elif opr=="//":   
        return f"floor division: {x1//x2}"
    else:
        return "Invalid operator"
x1=int(input("Enter first no: "))
x2=int(input("Enter second no: "))
opr=input("Enter any one operator from (+,-,*,/,%,**,//): ")
print(calci(x1,x2,opr))
```

# Fine Calculation

```python
books={"Name":["Harry Potter","Zomba"],"Cost":[100,200]}
limit=7
fine=0
fine_list=[]
fine_list1=[]
delay_days=int(input("How much delay to return a book? "))
for i in books["Cost"]:
    for j in range(delay_days+1):
        if j>limit:
            if i == fine:
                fine=i
            else:
                fine+=25
                fine_list.append(fine)
        else:
            fine=0
            fine_list.append(fine)
    fine_list1.append(fine_list[-1])
    books["Cost"]=fine_list1
print(books)
```
