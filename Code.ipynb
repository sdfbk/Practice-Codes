```python
import bs4,requests
from bs4 import BeautifulSoup as bs
url = "https://www.w3schools.com/"
response = requests.get(url,"html.parser")
soup = bs(response.text)
print(soup.prettify())
```

```python
l=[]
div1 = soup.find("div",class_="w3-hide-large w3-hide-medium")
div2 = div1.find_all("a")
for i in div2:
    l.append(i.text)
print(l)
```

```python
import re
text="Hello Good Morning all 123!!"
pattern = r"^([aeiou]|[AEIOU])"
match = re.search(pattern,text)
if match:
    print("Vowels")
else:
    print("Consonants")
```
